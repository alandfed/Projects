### Задача 1 (Парадокс инспектора)

Автобусы отходят от автостанции с 8:00 до 20:00. Первый автобус отходит ровно в 8:00. Затем интервалы между автобусами случайны, независимы и равновероятно равны либо 5-и, либо 10-и минутам. Будем считать, что за минуту на автостанцию приходит ровно один пассажир, и все пассажиры едут ближайшим автобусом.

Проведите $10^4$ экспериментов и с их помощью:

__а) [5 баллов]__ Постройте гистограмму количества автобусов, отошедших от автостанции за сутки. Похоже ли визуально распределение на нормальное?

__б) [10 баллов]__ Инспектор Тимон выбирает равновероятно один из всех автобусов отошедших от автостанции за сутки. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__в) [10 баллов]__ Инспектор Пумба приходит на автостанцию в случайный момент времени, равномерный от 8:00 до 20:00 и садится в первый пришедший автобус. Постройте гистограмму числа пассажиров на этом автобусе. Оцените математическое ожидание и дисперсию этого числа.

__г) [10 баллов]__ Как изменятся ответы на эти вопросы, если время между автобусами будет экспоненциально со средним в 10 минут?

__Пояснения:__ первый автобус можно не включать в выборку, Пумбу за пассажира можно не считать. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(555)
```

```python
plt.style.use('Solarize_Light2')
```

```python
buses, Timon, Pumba = list(), list(), list()
for _ in range(10 ** 4):
    time = 8 * 60
    prev = time

    breaks = []

    Pumba_bus = np.random.uniform(8 * 60, 20 * 60)

    while True:
        if time > Pumba_bus > prev:
            Pumba.append(time - prev)
        if time > 20 * 60:
            break
        prev = time
        time += np.random.choice([5, 10], 1, [0.5, 0.5])[0]
        
        breaks.append(time - prev)

    buses.append(len(breaks))
    Timon_bus = np.random.choice(breaks, 1)[0]
    Timon.append(Timon_bus)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(28, 10))

ax[0].hist(buses, bins=40)
ax[1].hist(Timon)
ax[2].hist(Pumba)

ax[0].set_title('Количество автобусов')
ax[1].set_title('Тимон - количество людей')
ax[2].set_title('Пумба - количество людей')

plt.setp(ax[0], ylabel='Частота')
plt.setp(ax[1], xlabel='Количество')

plt.show()
```

```python
print('Тимон. Равновероятный выбор автобуса\t\t\t E(x)\t D(x)')
print('\t\t\t\t\t\t\t', "%.2f" % np.mean(Timon), '\t', "%.2f" % np.var(Timon), '\n')

print('Пумба. Равномерное распределение для времени автобуса\t E(x)\t D(x)')
print('\t\t\t\t\t\t\t', "%.2f" % np.mean(Pumba), '\t', "%.2f" % np.var(Pumba), '\n')
```

Проанализируем полученные графики: 

- график распределения количества автобусов напоминает кривую нормального распределения. Так происходит из-за того, что "логика" нормального распределения также выполняется: чаще всего встречаются средние значения, и чем больше отклонение в стороны, тем меньше вероятность его повторения. Перефразируем: вероятность того, что все перерывы между автобусами будут равны только 5 или только 10 минутам меньше, чем вероятность того, что данные перерывы будут скомбинированы.

- график количества людей для автобуса Тимона показывает, что вероятность встречи 5 людей равна вероятности встречи 10 пассажиров в автобусе. Это объясняется тем, что выбор автобуса происходил равновероятно, как и длина перерыва между автобусами. Именно поэтому мы получили такое значения матожидания - 7.5 ((5 + 10) / 2).

- график количества людей для автобуса Пумбы показывает, что вероятность встречи 5 людей примерно в два раза меньше вероятности встречи 10 пассажиров в автобусе. Данную ситуацию можно объяснить устройством равномерного распределения: вероятность попадания в интервал в 10 минут больше, чем в 5-минутный (также в два раза). Также можно посмотреть на формулу вероятности попадания в интервал (a,b) (при равных а и b = 10 значение вероятности возрастает по сравнению с b = 5):

$$P(a < x < b) = (b - a) / (right - left)$$



```python
from scipy. stats import expon
```

```python
buses, Timon, Pumba = list(), list(), list()
for _ in range(10 ** 4):
    time = 8 * 60
    prev = time

    breaks = []

    Pumba_bus = np.random.uniform(8 * 60, 20 * 60)

    while True:
        if time > Pumba_bus > prev:
            Pumba.append(time - prev)
        if time > 20 * 60:
            break
        prev = time
        time += expon.rvs(scale=10)
        
        breaks.append(time - prev)

    buses.append(len(breaks))
    Timon_bus = np.random.choice(breaks, 1)[0]
    Timon.append(Timon_bus)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(28, 10))

ax[0].hist(buses, bins=70)
ax[1].hist(Timon, bins=50)
ax[2].hist(Pumba, bins=50)

ax[0].set_title('Количество автобусов')
ax[1].set_title('Тимон - количество людей')
ax[2].set_title('Пумба - количество людей')

plt.setp(ax[0], ylabel='Частота')
plt.setp(ax[1], xlabel='Количество')

plt.show()
```

```python
print('Тимон. Равновероятный выбор автобуса\t\t\t E(x)\t D(x)')
print('\t\t\t\t\t\t\t', "%.2f" % np.mean(Timon), '\t', "%.2f" % np.var(Timon), '\n')

print('Пумба. Равномерное распределение для времени автобуса\t E(x)\t D(x)')
print('\t\t\t\t\t\t\t', "%.2f" % np.mean(Pumba), '\t', "%.2f" % np.var(Pumba), '\n')
```

Проанализируем полученные графики: 

- график распределения количества автобусов теперь более рваный, так как перерыв определяется экспоненциальным распределением. Однако так как мы установили среднее значение 10, основные идеи сохраняются.

- график количества людей для автобуса Тимона теперь напоминает график плотности экспоненциального распределения, что правильно, так как теперь данный вид распределения решает, какой длины будет перерыв между автобусами, что равняется количеству людей в следующем. 

- график количества людей для автобуса Пумбы после преобразований напоминает комбинацию нормального и экспоненциального распределений, так как можно заметить волну графика, после чего идет график плотности экспоненциального распределения. Так происходит из-за того, что мы также равномерно выбираем время прибытия Пумбы, а для изменённого вида распределения устанавливаем среднее равное 10 (пик графика находится в значении 10).


Также из-за применения экспоненциального распределения разброс значений сильно вырос, выросла дисперсия.

- Если богатство посетителя кратно трём, то посетитель выигрывает франк с вероятностью $0.09$ и проигрывает один франк с вероятностью $0.91$

Арамис каждый раз выбирает казино А или казино Б равновероятно. Проведите $10^4$ симуляций эволюции благосостояния Атоса, Портоса и Арамиса.Каждая симуляция предусматривает $1000$ посещений казино. Все три игрока все деньги носят с собой и ни на что не тратят :)

__а) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Атоса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.

__в) [10 баллов]__ Постройте на графике $10$ случайных траекторий изменений богатства Арамиса, по горизонтали — номер посещени казино, по вертикали — богатство. На том же графике постройте усреднённую по всем экспериментам тракторию изменения богатства.
    

```python
def CasinoPortos(w):
    if w % 3 == 0:
        to_add = np.random.choice([-1, 1], size=1, p=[0.91, 0.09])[0]
    else:
        to_add = np.random.choice([-1, 1], size=1, p=[0.26, 0.74])[0]
    return to_add

def CasinoAtos(w):
    to_add = np.random.choice([-1, 1], size=1, p=[0.51, 0.49])[0]
    return to_add
```

```python
Atos = np.array([([1000] * 1000)] * 10**4)
Portos = np.array([([1000] * 1000)] * 10**4)
Aramis = np.array([([1000] * 1000)] * 10**4)

for j in range(10 ** 4):
    Atos_money = 1000
    Portos_money = 1000
    Aramis_money = 1000
    for i in range(1000):
        Atos_money += CasinoAtos(Atos_money)
        Atos[j][i] = Atos_money
        Portos_money += CasinoPortos(Portos_money)
        Portos[j][i] = Portos_money
        choice_Aramis = np.random.choice(["Atos_casino", "Portos_casino"], 1, [0.5, 0.5])[0]
        if choice_Aramis == "Atos_casino":
            Aramis_money += CasinoAtos(Aramis_money)
        elif choice_Aramis == "Portos_casino":
            Aramis_money += CasinoPortos(Aramis_money)
        Aramis[j][i] = Aramis_money
```

```python
fig, ax = plt.subplots(1, 3, figsize=(28, 8))

j_s = np.random.choice(np.arange(10000), size=10)
for j in j_s:
    ax[0].plot(Atos[j])
    ax[1].plot(Portos[j])
    ax[2].plot(Aramis[j])

ax[0].plot(np.mean(Atos, axis=0))
ax[1].plot(np.mean(Portos, axis=0))
ax[2].plot(np.mean(Aramis, axis=0))

ax[0].set_title('Атос')
ax[1].set_title('Портос')
ax[2].set_title('Арамис')

plt.setp(ax[0], ylabel='Деньги')
plt.setp(ax[1], xlabel='Номер похода в казино')

plt.show()
```

Данный парадокс заключается в следующем: комбинация двух проигрышных стратегий становится выигрышной стратегией.

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

```python
students = [0] * 17
i = 0
students_c, not_students_c = [], []

for n in range(80, 164, 5):
    students_marks_now = list()
    not_students_marks_now = list()
    for _ in range(10 ** 4):
        math = np.random.normal(60, 10)
        russian = np.random.normal(60, 10)
        if math + russian >= n:
            students[i] += 1
            students_marks_now.append([math, russian])
        else:
            not_students_marks_now.append([math, russian])

    students_marks_ = np.array(students_marks_now)
    students_c.append(np.corrcoef(students_marks_[::, 0], students_marks_[::, 1])[1][0])
    not_students_marks_ = np.array(not_students_marks_now)
    not_students_c.append(np.corrcoef(not_students_marks_[::, 0], not_students_marks_[::, 1])[1][0])
    i += 1
```

```python
fig, ax = plt.subplots(1, 3, figsize=(28, 8))

ax[0].plot(students)
ax[1].plot(students_c)
ax[2].plot(not_students_c)

ax[0].set_title('Прошедшие в УШЭ')
ax[1].set_title('Корреляция для прошедших')
ax[2].set_title('Корреляция для не прошедших')

plt.setp(ax[0], ylabel='Количество людей')
plt.setp(ax[1], ylabel='Значение корреляции')
plt.setp(ax[2], ylabel='Значение корреляции')
plt.setp(ax[1], xlabel='Проходной балл (80 + 5 * x)')

plt.show()
```

Проанализируем полученные графики:

- Число прошедших на обучение в УШЭ падает при увеличении порогового балла, причем при наименьшем пороге (равным 80) проходят все, а при максимальном (равном 160) - никто. Объясним: мы использовали нормальное распределение со средним 60 и отклонением 10, что означает следующее: два случайно сгенерированных числа таким распределением всегда будут в сумме давать 80 и никогда не будут давать 160. Поэтому количество прошедших уменьшается, когда пороговый балл растёт (вероятность того, что сумма двух значений будет давать необходимый балл падает).

- корреляция для прошедших на обучение УШЭ изменяется от 0 к -1, что показывает появление обратной корреляции, то есть присутствие зависимости между переменными. Значит, корреляция между баллами за русский и баллами за математику появляется при большем пороге, так как при низком пороге проходят все студенты (кто получили низкий балл за русский и высокий балл за математику, кто получили выоский балл за русский и низкий балл за математику, кто получили высокий балл за русский и высокий балл за математику), при такой выборке невозможно увидеть закономерность, а при высоком проходном балле проходит только группа студентов с высокими баллами, и тут возможно выделить закономерность.

- корреляция для не прошедших студентов на обучение УШЭ изменяется от -1 к 0, что показывает отсутствие закономености при высоком пороге и присутствие при низком. Значит, корреляция между баллами за русский и баллами за математику присутствует при низком пороге, так как в таких случаях не проходят студенты, которые сдали оба экзамена на низкие баллы (есть закономерность), а при увеличении порогового балла не проходит большая группа студентов (нет закономерности).
